```python
# -------------------------------------------------
# WITH/Concat of 50 unzipping "listings.csv.gz" files from Yang's MergedEdinburgh folder WITH CREATING THE Combined_description
# -------------------------------------------------

# * 0.Globs all *_listings.csv.gz files from the MergedEdinburgh folder and builds a numeric lookup (source_id 1…n) 
# based on chronological filename order
# * 1. Loops through each file, decompresses on the fly, reads only 4 columns (id, description, last_scraped, 
# neighborhood_overview), and skips any snapshot before 2019
# * 2. Step 1.Tags each row with its source_id so I can trace it back to the original .gz file
# * 3. Step 2.Fills missing description values with neighborhood_overview as a fallback
# * 4. Step 3.Creates a combined_description column by concatenating description + neighborhood_overview, handling NaNs 
# gracefully

# * 5. Vertically stacks all individual DataFrames into one df_merged
# * 6. Deduplicates on combined_description, keeping the first occurrence, while preserving a list of all source_ids 
# where that description appeared
# * 7.Saves the result as LLM_input_combined.csv — the direct input for the LLM-based POI extraction stage
# * 8.Exports a file_id_lookup.csv reference table mapping each source_id back to its original filename


import glob
import os
import pandas as pd
import re

#0.Globs all *_listings.csv.gz files from the MergedEdinburgh folder and builds a numeric lookup (source_id 1…n) based 
# on chronological filename order

# Path for gz files
folder_path = r"/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/MergedEdinburgh"

# Broader pattern — catches any gz file with "listings" in the name
path_pattern = os.path.join(folder_path, "*_listings.csv.gz")
all_files = glob.glob(path_pattern)

print(f"Files found: {len(all_files)}")
print("Actual folder contents:", os.listdir(folder_path)[:10])

dfs = []

# Before the loop — build a lookup from filename to numeric ID
all_files_sorted = sorted(all_files)
file_to_id = {os.path.basename(f): i for i, f in enumerate(all_files_sorted, start=1)}
# e.g. {"2019-03-14_listings.csv.gz": 1, "2019-06-22_listings.csv.gz": 2, ...}

#1. Loops through each file, decompresses on the fly, reads only 4 columns (id, description, last_scraped, 
# neighborhood_overview), and skips any snapshot before 2019
for filename in all_files:
    base_name = os.path.basename(filename)

    try:
        date_str = base_name[:10]
        try:
            year = int(date_str[:4])
        except ValueError:
            year = 9999

        if year >= 2019:
            df = pd.read_csv(
                filename,
                compression='gzip',
                usecols=['id', 'description', 'last_scraped', 'neighborhood_overview']
            )

           # 2. Step 1. Tags each row with its source_id so I can trace it back to the original .gz file
            df['source_id'] = file_to_id[base_name]

            # Step 2: Fill missing description with neighborhood_overview
            if 'neighborhood_overview' in df.columns:
                df['description'] = df['description'].fillna(df['neighborhood_overview'])

            # Step 3: Create combined_description
            df["combined_description"] = (
                df["description"].fillna("") + " " +
                df["neighborhood_overview"].fillna("")
            ).str.strip()

            dfs.append(df)
        else:
            print(f"  Skipped: {base_name} (year {year} < 2019)")

    except Exception as e:
        print(f"error processing file {base_name} : {e}")


# 5. Concatenate (Vertical stack)
if len(dfs) == 0:
    print("\n dfs is empty! Nothing to concatenate.")
    print("Check the printed filenames above — do they start with a year >= 2019?")
else:
    df_merged_cd = pd.concat(dfs, axis=0, ignore_index=True)
    print("Final VERTICAL stacked shape:", df_merged_cd.shape)

# 6. Collect all source_ids per unique description, then deduplicate
sources_cd = df_merged_cd.groupby("combined_description")["source_id"].apply(list).reset_index()
sources_cd.columns = ["combined_description", "source_ids"]

# Keep first occurrence — this row already has id, last_scraped, combined_description
df_unique_cd = df_merged_cd.drop_duplicates(subset=["combined_description"], keep="first")

 # Merge source_ids list back — df_unique already carries id and last_scraped from the first row
df_unique_cd = df_unique_cd[["id", "combined_description", "last_scraped"]].merge(
        sources_cd, on="combined_description")

# 7. Save
out_path = "/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/MergedEdinburgh/LLM_input_combined.csv"
df_unique_cd.to_csv(out_path, index=False)

rows, cols = df_unique_cd.shape
print(f"Original rows: {len(df_merged_cd)}, after removing duplicate descriptions: {len(df_unique_cd)}")
print(f"Saved to: {out_path}")
print(f"df_unique -> Rows: {rows}, Columns: {df_unique_cd.columns.tolist()}")

 
 #8.Exports a file_id_lookup.csv reference table mapping each source_id back to its original filename for traceability
lookup_path = os.path.join(folder_path, "file_id_lookup.csv")
pd.DataFrame(file_to_id.items(), columns=["filename", "source_id"]).to_csv(lookup_path, index=False)
print(f"File ID lookup saved to: {lookup_path}")
```

```python
# -------------------------------------------------
# Concat of 50 unzipping "listings.csv.gz" files
# from Yang's MergedEdinburgh folder — DESCRIPTION ONLY (no combined_description)
# -------------------------------------------------

import glob
import os
import pandas as pd
import re

# 0. Globs all *_listings.csv.gz files and builds a numeric lookup (source_id 1…n)
folder_path = r"/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/MergedEdinburgh"

path_pattern = os.path.join(folder_path, "*_listings.csv.gz")
all_files = glob.glob(path_pattern)

print(f"Files found: {len(all_files)}")
print("Actual folder contents:", os.listdir(folder_path)[:10])

dfs = []

# Before the loop — build a lookup from filename to numeric ID
all_files_sorted = sorted(all_files)
file_to_id = {os.path.basename(f): i for i, f in enumerate(all_files_sorted, start=1)}

# 1. Loop: decompress, read 4 columns, skip years < 2019
for filename in all_files:
    base_name = os.path.basename(filename)

    try:
        date_str = base_name[:10]
        try:
            year = int(date_str[:4])
        except ValueError:
            year = 9999

        if year >= 2019:
            df = pd.read_csv(
                filename,
                compression='gzip',
                usecols=['id', 'description', 'last_scraped', 'neighborhood_overview']
            )

            # 2. Tag each row with its source_id
            df['source_id'] = file_to_id[base_name]

            # 3. Fill missing description with neighborhood_overview
            if 'neighborhood_overview' in df.columns:
                df['description'] = df['description'].fillna(df['neighborhood_overview'])

            dfs.append(df)
        else:
            print(f"  Skipped: {base_name} (year {year} < 2019)")

    except Exception as e:
        print(f"error processing file {base_name} : {e}")


# 4. Concatenate (Vertical stack)
if len(dfs) == 0:
    print("\n dfs is empty! Nothing to concatenate.")
    print("Check the printed filenames above — do they start with a year >= 2019?")
else:
    df_merged = pd.concat(dfs, axis=0, ignore_index=True)
    print("Final VERTICAL stacked shape:", df_merged.shape)

    # 5. Collect all source_ids per unique description, then deduplicate
    sources_d = df_merged.groupby("description")["source_id"].apply(list).reset_index()
    sources_d.columns = ["description", "source_ids"]

    # Keep first occurrence — this row already has id, last_scraped, description
    df_unique = df_merged.drop_duplicates(subset=["description"], keep="first")

    # Merge source_ids list back
    df_unique = df_unique[["id", "description", "last_scraped"]].merge(
        sources_d, on="description"
    )

    # 6. Save
    out_path = "/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/MergedEdinburgh/LLM_input_description_only.csv"
    df_unique.to_csv(out_path, index=False)

    rows, cols = df_unique.shape
    print(f"Original rows: {len(df_merged)}, after removing duplicate descriptions: {len(df_unique)}")
    print(f"Saved to: {out_path}")
    print(f"df_unique -> Rows: {rows}, Columns: {df_unique.columns.tolist()}")

```

```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt

# Path to the single CSV file
csv_path = "/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/MergedEdinburgh/LLM_input.csv"

fname = os.path.basename(csv_path)

# Read the CSV
df = pd.read_csv(csv_path)

# Derive year_month_* from last_scraped (e.g. '2023-01-01' -> '2023-01')
# If new_id is '2023-01-01', first 7 chars give '2023-01'
df["year_month"] = df["last_scraped"].astype(str).str.slice(0, 7)

# Count number of listings (id) per year_month
counts = (
    df.dropna(subset=["year_month"])
      .groupby("year_month")["id"]
      .nunique()  # unique new_id per month; use .count() for raw rows
      .reset_index(name="id_count")
)


# Sort and plot
counts["year_month_dt"] = pd.to_datetime(counts["year_month"], format="%Y-%m") # dt is necessary other wish the x-label will look like 2025_01_19 00:00;00)
counts = counts.sort_values("year_month_dt")

labels = counts["year_month"].tolist()
values = counts["id_count"].tolist()

plt.figure(figsize=(8, 4))
x_pos = range(len(labels))
plt.plot(x_pos, values, marker="o")
plt.xticks(x_pos, labels, rotation=90, ha="right")
plt.xlabel("Year-Month (from last_scraped)")
plt.ylabel("Number of listings")
plt.title(f"Count of listings by Year_Month (from last_scraped)")
plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------------------------------------------------------------------------------------------------
# Average Length of description column plotted against filename in the Unzipping 50 listings.csv.gz files from Yang's MergedEdinburgh folder
# ----------------------------------------------------------------------------------------------------------------------------------------

# This python script reads all listing .csv files from Yang's MergedEdinburgh folder and finds the send column named "name" 
# where the listing description is inlcuded . It counts the number of characters in that string , finds the mean for each file 
# and then plots the name of the files against the obeserved means, indicating in green which years we have longer listing descriptions

import glob
import os

import pandas as pd
import matplotlib.pyplot as plt

# Folder where all listings CSVs from the MergedEdinburgh download are stored
DATA_DIR = "/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/MergedEdinburgh/unzipped_csvs"  # /path/to/your/csvs

# Pattern for files like "2024-05-21_listings.csv"
pattern = os.path.join(DATA_DIR, "202*-*-*_listings.csv")  # * added to capture all combinations, if I want only September I replace 202*-10-*_listings.csv

file_paths = sorted(glob.glob(pattern))

file_labels = []
char_counts = []

for path in file_paths:
    # Short label for x-axis (just the filename)
    fname = os.path.basename(path)
    file_labels.append(fname)

 # Read the CSV INSIDE the loop (This cannot be MISSING and cannot be after the if below)
    df = pd.read_csv(path)

    if "description" not in df.columns:
        print(f"Warning: 'description' column not found in {fname}, skipping.")
        continue
 

    # Convert to string (in case of NaNs or non-string types)
    name_str = df["description"].astype(str)

    # Count characters per row
    char_count = name_str.str.len().mean() 
    char_counts.append(char_count)


    # print immediately if above threshold
    if char_count > 55:
        print("Above 55 characters:", fname, char_count)

# Build a color list: green for file-labels with char_count >55, else default blue
colors = ["green" if c > 55 else "blue" for c in char_counts]

# Use positions 0..N-1 as x, then label ticks with filenames
x_pos = range(len(file_labels))

# Sanity check
print("file_labels:", file_labels)
print("char_counts:", char_counts)

print("DATA_DIR exists?", os.path.exists(DATA_DIR))
print("pattern =", pattern)
print("Found files:", glob.glob(pattern))

# List ALL csv files in the directory to see what's actually there
print("\nAll CSV files in dir:")
print(glob.glob(os.path.join(DATA_DIR, "*.csv")))


x_pos = range(len(file_labels))

# Simple Scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(x_pos, char_counts, c=colors, marker="o", zorder=2)
plt.xticks(x_pos, file_labels, rotation=45, ha="right")
plt.xlabel("CSV file")
plt.xlabel("CSV file")
plt.ylabel("Average number of characters in 'description' column")
plt.title("MergedEdinburgh/gz:Description length vs CSV file (green > 55)")

plt.tight_layout()
plt.show()

# Simple line plot
plt.figure(figsize=(14, 5))
plt.plot(file_labels, char_counts, marker="o")
plt.xticks(rotation=45, ha="right")
plt.xlabel("CSV file")
plt.ylabel("Average number of characters in 'description' column")
plt.title("MergedEdinburgh/gz:Description length vs CSV file")
plt.tight_layout()
plt.show()

```

```python

# ----------------------------------------------------------------------------------------------------------------------------------------
# Missing values: which columns(variables) and percentages of missing values plotted (Heatmap) against file name 
# in the Unzipping 50 listings.csv.gz files from Yang's MergedEdinburgh folder
# ----------------------------------------------------------------------------------------------------------------------------------------


# This script reads all listing.csv  originally downloaded from Yang's MergedEdinburgh folder, now organised in 
# a new subfolder called unzipped_csvs. It identifies 42 colums of 75 having missing values and their percentange for each column.

# The script plots missing % for all CSV files for the selected 42 columns only.

# WHICH COLUMS HAVE MISSING VALUES: We observe  42 columns: description_pct , neighborhood_overview_pct, picture_url_pct, host_name_pct, 
#host_since_pct,	host_location_pct ,	host_about_pct ,	host_response_time_pct ,	host_response_rate_pct, host_acceptance_rate_pct, 	
#host_is_superhost_pct ,	host_thumbnail_url_pct	, host_picture_url_pct	, host_neighbourhood_pct, 	host_listings_count_pct	, 
#host_total_listings_count_pct, 	host_verifications_pct,	host_has_profile_pic_pct, 	host_identity_verified_pct, 	
#neighbourhood_pct, neighbourhood_group_cleansed_pct, bathrooms_pct, 	bathrooms_text_pct, 	bedrooms_pct,	beds_pct, price_pct,
#minimum_minimum_nights_pct, minimum_maximum_nights_pct, 	maximum_maximum_nights_pct, 	minimum_nights_avg_ntm_pct, 	
#maximum_nights_avg_ntm_pct ,	calendar_updated_pct,	has_availability_pct , estimated_revenue_l365d_pct, first_review_pct,	
#last_review_pct,	review_scores_rating_pct,	review_scores_accuracy_pct,	review_scores_cleanliness_pct,	review_scores_checkin_pct,
#review_scores_communication_pct,	review_scores_location_pct,	review_scores_value_pct,	license_pct, reviews_per_month_pct 

import glob
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


DATA_DIR = "/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/MergedEdinburgh/unzipped_csvs"
pattern = os.path.join(DATA_DIR, "202*-*-*_listings.csv")

file_paths = sorted(glob.glob(pattern))

results = []

for path in file_paths:
    fname = os.path.basename(path)
    
    # Read the CSV
    df = pd.read_csv(path)
    
    # Count missing values per column
    missing_counts = df.isnull().sum()
    
    # Total rows
    total_rows = len(df)
    
   # PERCENTAGE missing per column
    missing_pct = (missing_counts / total_rows * 100).round(2)

    # Create row for each file
    row = {"filename": fname, "total_rows": total_rows}
    
    # Add percentange and count (optional) of missing values for each column
    for col in df.columns:
        #row[f"{col}_count"] = missing_counts[col]
        row[f"{col}_pct"] = missing_pct[col]
    
    results.append(row)

# Convert to DataFrame
all_missing = pd.DataFrame(results)

# Reorder columns: filename, total_rows, then pct columns
cols = ["filename", "total_rows"]
for col in df.columns:  # df from last file, but all files have same columns
    cols.append(f"{col}_pct")
all_missing = all_missing[cols]

print("Missing values table (first 75 columns):")
print(all_missing.iloc[:, :75].round())

# ---- NEW: list all columns that have missing values in ANY file ----
# Drop the metadata columns
pct_only = all_missing.drop(columns=["filename", "total_rows"])

# Keep only columns where at least one file has missing % > 0
cols_with_missing = pct_only.columns[(pct_only > 0).any()].tolist()

# Strip the '_pct' suffix to get original column names
cols_with_missing_base = [c.replace("_pct", "") for c in cols_with_missing]

print("\nColumns that have missing values in at least one file:")
for c in cols_with_missing_base:
    print(c)
# --------------------------------------------------------------------

# Save full table
all_missing.to_csv("/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/Python/missing_values_table.csv", index=False)
print("\nFull table saved to missing_values_table.csv in this path /Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/Python/missing_values_table.csv")

# ------------------------------------------------------------------
#                             PLOTTING 
# ------------------------------------------------------------------

# list of base column names I care about. An example of 3 valiables (columns) was given below
selected_cols = ['price', 'neighborhood_overview','host_neighbourhood', 'neighbourhood', 'neighbourhood_cleansed', 'neighbourhood_group_cleansed']

# convert to the _pct column names present in all_missing
selected_pct_cols = [f"{c}_pct" for c in selected_cols]

# Keep all files (or slice, e.g. [:10] for first 10)
plot_df = all_missing[["filename"] + selected_pct_cols]

# Melt data for plotting (one row per file/column pair)
plot_data = plot_df.melt(
    id_vars=["filename"],
    var_name="column",
    value_name="missing_pct"
)

# Extract column name from pct column names, i.e., remove "_pct" suffix in labels
plot_data["column"] = plot_data["column"].str.replace("_pct$", "", regex=True)

# Create heatmap 
plt.figure(figsize=(10, 8))

# Pivot to 2D table: rows=filename, columns=column, values=missing_pct
heatmap_data = plot_data.pivot(index="filename", columns="column", values="missing_pct")

sns.heatmap(
    heatmap_data,
    annot=True,
    cmap="YlOrRd",
    fmt=".1f",
    cbar_kws={'label': 'Missing %'}
)

plt.title("Missing Values % - All Files (Selected Columns)", fontsize=14)
plt.xlabel("Columns")
plt.ylabel("Filename")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

output_dir = "/Users/xrysa/Desktop/heatmaps"
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, "missing_values_selected_columns_heatmap.png")

plt.savefig(output_path, dpi=300, bbox_inches='tight')
plt.show()

print("Saving to:", os.getcwd())
print("Heatmap saved to:", output_path)
```

```python
# ----------------------------------------------------------------------------------------------------------------------------------------
# HOW MANY LISTINGS SCRAPED THE SAME DATE? What is the Average Length of description column PER LAST_SCRAPED DATE plotted against file 
# in the analysis_descriptions..csv file?
# ----------------------------------------------------------------------------------------------------------------------------------------

# This python script reads the aLLM_input_without_Teacher_rows.csv file and and finds the send column 
# named "description" where the listing description is inlcuded . It counts the number of characters in that string. 
# It finds the string having the same date in the 'last-scraped' column, calcluate the mean of character count per same date (year-month-*). 
# Then it plots the (year-month-*)  against the count of the characters in the description.


import glob
import os

import pandas as pd
import matplotlib.pyplot as plt

  # Read the CSV
df = pd.read_csv("/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/YangGithub/analysis_descriptions.csv")



# specify the column where I want to find duplicates: "last-scraped"
col = "last_scraped"

# Boolean mask: True for rows where the value in col is duplicated
mask_dups = df.duplicated(subset=[col], keep=False)

# Subset of rows with duplicate values in that column
dup_rows = df[mask_dups]
print(f"Total rows in dataset: {len(df)}")
print(f"Rows with duplicate {col}: {len(dup_rows)}")
#print(dup_rows)

# ------------------------------------------------------------------
# For those duplicate dates, compute average number of characters
# in the 'description' column per last_scraped date
# ------------------------------------------------------------------

if "description" not in dup_rows.columns:
    raise KeyError("'description' column not found in the DataFrame")


# Ensure description is string
dup_rows["description"] = dup_rows["description"].astype(str)
    
# Length of each description
dup_rows["desc_len"] = dup_rows["description"].str.len()


# Aggregate subsets i.e., average description length per last_scraped (grouped by)
agg = dup_rows.groupby(col)["desc_len"].mean().reset_index()

# Count rows per subset (how many rows in each duplicate group):
counts_per_group = dup_rows.groupby(col).size().reset_index(name="count")
print(f"Number of unique {col} with duplicates: {len(agg)}")
print("Rows per duplicate group (first 5):")
print(counts_per_group)


file_labels = agg[col].astype(str).tolist()     # x-axis labels (dates if Option 2 or Year_Month if Option 1 selected)
char_counts = agg["desc_len"].tolist()          # y-axis values

# Plot
plt.figure(figsize=(14, 5)) # I added a bit of space in the x-axis
x_pos = range(len(file_labels))
plt.plot(x_pos, char_counts, marker="o")
plt.xticks(x_pos, file_labels, rotation=90, ha="right")
plt.xlabel("The same last_scraped date") 
plt.ylabel("Average number of characters in 'description'")
plt.title("GitHub_Analysis:Average length of lisitng description for subgroups with the same last_scraped dates") 
plt.show()



```

```python
# ----------------------------------------------------------------------------------------------------------------------------------------
# HOW MANY LISTINGS SCRAPED WITHIN THE SAME YEAR_MONTH? What is the Average Length of description column GROUPED PER YEAR_MONTH plotted against file in the analysis_descriptions.csv file 
# from Yang's GitHub LLM/data folder? 
# ----------------------------------------------------------------------------------------------------------------------------------------

# This python script reads the analysis_descriptions.csv file from Yang's LLM/data folder in her GitHub and and finds the column 
# named "description" where the listing description is inlcuded, Group the rows having the same Year_Month (all scrapped dates grouped together)
# It calcluate the mean of character count per same date (year-month-*). 
# Then it plots the (year-month-*)  against the count of the characters in the description.

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV
df = pd.read_csv("/Users/xrysa/Desktop/MSc_in_Urban_Analytics/Group_Project/DATA/YangGithub/analysis_descriptions.csv")

# specify the column where I want to find duplicates: "last-scraped"
col = "last_scraped"

# Boolean mask: True for rows where the value in col is duplicated
mask_dups = df.duplicated(subset=[col], keep=False)

# Subset of rows with duplicate values in that column
dup_rows = df[mask_dups]
print(f"Total rows in dataset: {len(df)}")
print(f"Rows with duplicate {col}: {len(dup_rows)}")

if "description" not in dup_rows.columns:
    raise KeyError("'description' column not found in the DataFrame")

# Ensure description is string
dup_rows["description"] = dup_rows["description"].astype(str)

# Length of each description
dup_rows["desc_len"] = dup_rows["description"].str.len()

# Extract year-month from last_scraped (e.g., "2023-05-15" → "2023-05")
dup_rows["year_month"] = pd.to_datetime(dup_rows[col]).dt.strftime("%Y-%m")

# Aggregate: average description length per year-month (only for duplicate dates)
agg = dup_rows.groupby("year_month")["desc_len"].mean().reset_index()

# Count rows per group
counts_per_group = dup_rows.groupby("year_month").size().reset_index(name="count")
print(f"Number of unique year-month with duplicates: {len(agg)}")
print("Rows per duplicate year-month (first 5):")
#print(counts_per_group.head())
print(counts_per_group)

file_labels = agg["year_month"].tolist()        # x-axis labels (e.g., "2023-05")
char_counts = agg["desc_len"].tolist()          # y-axis values

# Plot
plt.figure(figsize=(14, 5))
x_pos = range(len(file_labels))
plt.plot(x_pos, char_counts, marker="o")
plt.xticks(x_pos, file_labels, rotation=45, ha="right")
plt.xlabel("Year-Month (last_scraped)")
plt.ylabel("Average number of characters in 'description'")
plt.title("GitHub_Analysis:Average length of lisitng description for subgroups with the same Year_Month")
plt.tight_layout()
plt.show()
```
